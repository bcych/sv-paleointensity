### Get cluster numbers (-1 is the label assigned to states with no cluster)

```python
labels = cluster.cluster_structures(cluster_df)
```

### Get pairs of states to calculate barriers between. 
States with "r" indicate a rotation. States with "m" indicate a helicity switching. 
Whether a rotation or a helicity switching occurs depends on the last state, not the first.
repr_keys is a dictionary of unique ids. repr_pairs is a dictionary of key pairs for barrier calculations. Sometimes this causes key-value clashes where a helicity swap is to a rotated state, so you might see for example an s0 -> s0r2 and an s0r1 -> s0r2 where the second is the rotation, a bit confusing but can be worked out from context.

```python
repr_pairs,repr_keys = cluster.get_barrier_endpoints(cluster_df,labels)
repr_keys, repr_pairs
```

### This is how you list the states you're pairing:

```python
for pair in repr_pairs:
    print(repr_keys[pair[0]],'->',repr_keys[pair[1]])
```
